```python
import matplotlib.pyplot as plt
import numpy as np

with open("./nohup.out", 'rb') as f:
    losses = []
    line = f.readline()
    while line:
        if not line.startswith("Reward"):
            line = f.readline()
            continue
        loss = line.split("Reward: ")[1].strip()
        loss = float(loss)
        losses.append(loss)
        line = f.readline()
    print len(losses)
    losses = np.array(losses)
    plt.plot(losses)
    plt.xlabel("iteration / 10")
    plt.ylabel("loss")
    plt.show()
```

```python
import numpy as np


plt.plot(np.where(losses > 300)[0].tolist())
```

```python
temp = np.where(losses > 300)[0].tolist()
temp2 = [temp[i]-temp[i-1] for i,j in enumerate(temp)][1:]
plt.plot(temp2)
plt.xlabel("gap over 300s")
plt.ylabel("reward")
plt.show()
```

```python
temp3 = map(lambda x:x/200.0, [sum(losses[i:i+200]) for i,j in enumerate(losses)][:-200])
plt.plot(temp3)
plt.xlabel("Average every 100")
plt.ylabel("reward")
plt.show()
```
